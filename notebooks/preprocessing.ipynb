```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/solar_AL.txt", header=None)
data.shape
```

```python
data["sequenceId"] = np.ones(data.shape[0]).astype(int)
data["itemPosition"] = np.arange(data.shape[0]).astype(int)
```

```python
data = data[["sequenceId", "itemPosition"] + list(range(data.shape[1]-2))]
```

```python
data.to_csv("../data/solar_AL.txt", index=None)
```
